#### To set up the data taken from PGAdmin as a CSV, mostly about cleaning other data, not the arrest records

```python
import pandas
import numpy
import re
```

```python
def main(datafile):
    dataset = pandas.read_csv(datafile, quotechar = '"', low_memory = False)
    
    dataset = prepare_basic(dataset)
    dataset = prepare_races(dataset)
    
    dataset = calc_resrace(dataset)
    dataset = mugshots_ator(dataset)
    dataset = calc_police(dataset)
    dataset = calc_policevres(dataset)
    dataset = calc_raceinc(dataset)
    dataset = ucr_weight(dataset)
    
    return dataset
```

```python
def prepare_basic(dataset):
    # Remove any mugshots without a county identified
    dataset = dataset[dataset['county_no'].notnull()]
    
    for date_var in ['date', 'download_date', 'birthday', 'views_date']:
        dataset[date_var] = pandas.to_datetime(dataset[date_var], coerce = True, utc = True)

    # dataset['days_online'] = dataset['download_date'] - dataset['date']
    
    # Removing minors without removing people of unknown age
    dataset = dataset.drop(dataset.index[dataset['age'] < 18])
    dataset['age_sq'] = dataset['age'] ** 2
    
    dataset = pandas.concat([dataset, pandas.get_dummies(dataset['gov_party'])], axis = 1)
    dataset = dataset.rename(columns = {'Democrat': 'gov_dem', 'Republican': 'gov_repub', \
                                        'Independent': 'gov_indep'})
    
    return dataset
```

```python
def prepare_races(dataset):    
    races = ['white', 'am_ind', 'asian', 'east_asian', 'hispanic', 'black']

    # [:] to pass by value, to keep hispanic in first list
    races_nohisp = races[:]
    races_nohisp.remove('hispanic')

    # Making multi-racial var, counting hispanic as its own race
    dataset.loc[dataset[races].sum(axis = 1) == 1, 'multi_race_whisp'] = 0
    dataset.loc[dataset[races].sum(axis = 1) > 1, 'multi_race_whisp'] = 1
    
    # Multi-racial, ignoring hispanic
    dataset.loc[dataset[races_nohisp].sum(axis = 1) == 1, 'multi_race_nohisp'] = 0
    dataset.loc[dataset[races_nohisp].sum(axis = 1) > 1, 'multi_race_nohisp'] = 1

    # Making reference group any of the other races found, ref. is overwritten if the race of interest is found
    for race in races:
        # If any race was found, make reference 0
        dataset.loc[dataset['multi_race_whisp'] == 0, race + '_only'] = 0
        dataset.loc[dataset[race] == 1, race + '_only'] = 1
        # If multi-racial
        dataset.loc[dataset['multi_race_whisp'] == 1, race + '_only'] = 0
        dataset.loc[dataset['multi'] == 1, race + '_only'] = 0

    # To ignore hispanicity, not consider it a race
    for race in races_nohisp:
        dataset.loc[dataset['multi_race_nohisp'] == 0, race + '_only_ighisp'] = 0
        dataset.loc[dataset[race] == 1, race + '_only_ighisp'] = 1
        # If multi-racial
        dataset.loc[dataset['multi_race_nohisp'] == 1, race + '_only_ighisp'] = 0
        dataset.loc[dataset['multi'] == 1, race + '_only_ighisp'] = 0
        
    dataset.loc[dataset['black_only_ighisp'] == 1, 'black_not_white'] = 1
    dataset.loc[dataset['white_only_ighisp'] == 1, 'black_not_white'] = 0

    return dataset
```

```python
def calc_resrace(dataset):
    dataset.loc[dataset['total_residents'] != 0, 'b_residents_percent'] = (dataset['b_residents'] / dataset['total_residents']) * 100
    dataset.loc[dataset['total_residents'] != 0, 'w_residents_percent'] = (dataset['w_residents'] / dataset['total_residents']) * 100
    dataset.loc[dataset['total_residents'] != 0, 'bw_residents_pdiff'] = dataset['b_residents_percent'] - dataset['w_residents_percent']
    dataset.loc[dataset['w_residents'] != 0, 'bw_residents_ratio'] = dataset['b_residents'] / dataset['w_residents']
    
    dataset['nonw_residents_percent'] = 100 - dataset['w_residents_percent']
    dataset['nonw_residents_percent_sq'] = dataset['nonw_residents_percent'] ** 2
    
    dataset['b_residents_percent_sq'] = dataset['b_residents_percent'] ** 2
    dataset['w_residents_percent_sq'] = dataset['w_residents_percent'] ** 2
    
    return dataset
```

```python
def calc_police(dataset):
    # filling with zeros to compute sum, but prob don't want to assume NA usually means 0 for other measures
    dataset['w_officers_zeros'] = dataset['w_officers'].fillna(0)
    dataset['b_officers_zeros'] = dataset['b_officers'].fillna(0)
    dataset['a_officers_zeros'] = dataset['a_officers'].fillna(0)
    dataset['n_officers_zeros'] = dataset['n_officers'].fillna(0)
    dataset['ai_officers_zeros'] = dataset['ai_officers'].fillna(0)
    dataset['mult_officers_zeros'] = dataset['mult_officers'].fillna(0)
    dataset['u_officers_zeros'] = dataset['u_officers'].fillna(0)

    # computing officer dems
    dataset['total_officers_calc'] = dataset['w_officers'] + dataset['b_officers'] \
        + dataset['a_officers'] + dataset['n_officers'] + dataset['ai_officers'] \
        + dataset['mult_officers'] + dataset['u_officers']

    dataset.loc[dataset['swnauthemp'] != 0, 'w_officers_percent'] = (dataset['w_officers'] / dataset['swnauthemp']) * 100
    dataset.loc[dataset['total_officers_calc'] != 0, 'w_officers_percent_calc'] = (dataset['w_officers'] / dataset['total_officers_calc']) * 100
    dataset.loc[dataset['total_officers_calc'] != 0, 'b_officers_percent_calc'] = (dataset['b_officers'] / dataset['total_officers_calc']) * 100
    dataset.loc[dataset['w_officers_percent_calc'] != 0, 'bw_officers_pdiff'] = dataset['b_officers_percent_calc'] - dataset['w_officers_percent_calc']
    dataset.loc[dataset['w_officers_percent_calc'] != 0, 'bw_officers_ratio'] = dataset['b_officers_percent_calc'] / dataset['w_officers_percent_calc']
    
    dataset['male_officers_percent'] = (dataset['male_officers'] / dataset['totgender']) * 100
    
    return dataset
```

```python
# Arrests to Residents Ratio
def mugshots_ator(dataset):
    counties = dataset.groupby('county_no')
    for county_no in counties.groups:
        dataset.loc[dataset['county_no'] == county_no, 'b_arrests'] = counties.get_group(county_no)['black_only'].sum()
        dataset.loc[dataset['county_no'] == county_no, 'w_arrests'] = counties.get_group(county_no)['white_only'].sum()
        dataset.loc[dataset['county_no'] == county_no, 'total_arrests'] = float(counties.get_group(county_no)['arrest_id'].count())

    dataset.loc[dataset['b_residents'] != 0, 'b_ator'] = dataset['b_arrests'] / dataset['b_residents']
    dataset.loc[dataset['w_residents'] != 0, 'w_ator'] = dataset['w_arrests'] / dataset['w_residents']
    dataset['bminusw_ator'] = dataset['b_ator'] - dataset['w_ator']
    dataset['bdivw_ator'] = dataset['b_ator'] / dataset['w_ator']

    dataset.loc[dataset['total_arrests'] != 0, 'b_arrests_percent'] =  dataset['b_arrests'] / dataset['total_arrests']
    dataset.loc[dataset['total_arrests'] != 0, 'w_arrests_percent'] =  dataset['w_arrests'] / dataset['total_arrests']
#    dataset['b_aminusr'] = dataset['b_arrests_percent'] - dataset['b_residents_percent']
#    dataset['w_aminusr'] = dataset['w_arrests_percent'] - dataset['w_residents_percent']
    
    return dataset
```

```python
def calc_policevres(dataset):
    dataset.loc[dataset['w_residents_percent'] != 0, 'wofficers_minusres'] =  dataset['w_officers_percent_calc'] - dataset['w_residents_percent']
    dataset.loc[dataset['b_residents_percent'] != 0, 'bofficers_minusres'] =  dataset['b_officers_percent_calc'] - dataset['b_residents_percent']
    dataset.loc[dataset['w_residents_percent'] != 0, 'wofficers_divres'] =  dataset['w_officers_percent_calc'] / dataset['w_residents_percent']
    dataset.loc[dataset['b_residents_percent'] != 0, 'bofficers_divres'] =  dataset['b_officers_percent_calc'] / dataset['b_residents_percent']
    return dataset
```

```python
def calc_raceinc(dataset):
    dataset.loc[dataset['binc_total'] != 0, 'binc_under10_percent'] = dataset['binc_under10'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_10to15_percent'] = dataset['binc_10to15'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_15to20_percent'] = dataset['binc_15to20'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_20to25_percent'] = dataset['binc_20to25'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_25to30_percent'] = dataset['binc_25to30'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_30to35_percent'] = dataset['binc_30to35'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_35to40_percent'] = dataset['binc_35to40'] / dataset['binc_total']
    dataset.loc[dataset['binc_total'] != 0, 'binc_under40'] = dataset['binc_under10'] + dataset['binc_10to15'] \
        + dataset['binc_15to20'] + dataset['binc_20to25'] + dataset['binc_25to30'] \
        + dataset['binc_30to35'] + dataset['binc_35to40']
    dataset.loc[dataset['binc_total'] != 0, 'binc_under40_percent'] = dataset['binc_under40'] / dataset['binc_total']

    dataset.loc[dataset['winc_total'] != 0, 'winc_under10_percent'] = dataset['winc_under10'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_10to15_percent'] = dataset['winc_10to15'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_15to20_percent'] = dataset['winc_15to20'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_20to25_percent'] = dataset['winc_20to25'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_25to30_percent'] = dataset['winc_25to30'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_30to35_percent'] = dataset['winc_30to35'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_35to40_percent'] = dataset['winc_35to40'] / dataset['winc_total']
    dataset.loc[dataset['winc_total'] != 0, 'winc_under40'] = dataset['winc_under10'] + dataset['winc_10to15'] \
        + dataset['winc_15to20'] + dataset['winc_20to25'] + dataset['winc_25to30'] \
        + dataset['winc_30to35'] + dataset['winc_35to40']
    dataset.loc[dataset['winc_total'] != 0, 'winc_under40_percent'] = dataset['winc_under40'] / dataset['winc_total']

    dataset['bwinc_under40_pdiff'] = dataset['binc_under40_percent'] - dataset['winc_under40_percent']
    
    return dataset
```

```python
def ucr_weight(dataset):
    # To weight underrepresented counties more (using monthly UCR data)
    for county_no in dataset['county_no'].unique():
        monthly_county_arrests = dataset[dataset['county_no'] == county_no]['arrest_id'].count() / 12
        dataset.loc[dataset['county_no'] == county_no, 'ucr_weight_raw'] = dataset['total_arrests'] / monthly_county_arrests

    dataset['ucr_weight_normalized'] = dataset['ucr_weight_raw'] / dataset['ucr_weight_raw'].sum()
    return dataset
```

```python
# UCR data at the county level
def ucr_ator(dataset):
    dataset['total_arrests'] = dataset['w_arrests'].fillna(0) + dataset['w_arrests'].fillna(0) \
        + dataset['a_arrests'].fillna(0) + dataset['i_arrests'].fillna(0)
    
    dataset.loc[dataset['total_arrests'] != 0, 'b_arrests_percent'] =  dataset['b_arrests'].astype(float) / dataset['total_arrests'].astype(float)
    dataset.loc[dataset['total_arrests'] != 0, 'w_arrests_percent'] =  dataset['w_arrests'].astype(float) / dataset['total_arrests'].astype(float)

    dataset.loc[dataset['b_residents'] != 0, 'b_ator'] = (dataset['b_arrests'].astype(float) / dataset['b_residents'].astype(float))
    dataset.loc[dataset['w_residents'] != 0, 'w_ator'] = (dataset['w_arrests'].astype(float) / dataset['w_residents'].astype(float))
    
    dataset['bminusw_ator'] = dataset['b_ator'] - dataset['w_ator']
    dataset['bdivw_ator'] = dataset['b_ator'] / dataset['w_ator']
    
    return dataset
```
